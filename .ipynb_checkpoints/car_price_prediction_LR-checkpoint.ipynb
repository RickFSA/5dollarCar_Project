```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error 
```

__Build base model__

```python
numeric_data = car[['year', 'engine_hp', 'engine_cylinders', 'number_of_doors', 'highway_mpg',
               'city_mpg', 'price']]
```

```python
X = numeric_data.drop('price', axis=1)
y = numeric_data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.2, random_state=1)

steps = [('scaler', StandardScaler()), ('mod', LinearRegression())]
pipe = Pipeline(steps)

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
print(f'Base model accuracy: {pipe.score(X_test, y_test):.3f}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, y_pred))}')
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
```

### __1st cross validation score__

```python
cross_val_score(pipe, X, y, cv=10).mean()
```

Our base-model's accuracy is 45%. We can proceed to next step for comparison

```python
clf = Ridge()

coefs = []
errors = []
ridge_alphas = np.logspace(-6, 6, 200)

for a in ridge_alphas:
    clf.set_params(alpha = a)
    clf.fit(X2, y2)
    coefs.append(clf.coef_)
    errors.append(mean_squared_error(clf.coef_, lreg.coef_))
```

```python
plt.figure(figsize = (20, 6))

plt.subplot(121)
ax = plt.gca()
ax.plot(ridge_alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge coefficients as a function of the regularisation')
plt.axis('tight')

plt.subplot(122)
ax = plt.gca()
ax.plot(ridge_alphas, errors)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('error')
plt.title('Coefficient error as a function of the regularisation')
plt.axis('tight')

plt.show()
```

```python
optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(X2, y2)
print('Alpha:', optimal_ridge.alpha_)
print('Score:', optimal_ridge.score(X2, y2))
```

```python
lasso_alphas = np.logspace(-6, 6, 200)
optimal_lasso = LassoCV(alphas=lasso_alphas, cv=10)
optimal_lasso.fit(X2, y2)
print('Alpha:', optimal_lasso.alpha_)
print('Score:', optimal_lasso.score(X2, y2))
```

```python
included = []
best = {'feature': '', 'r2': 0, 'ar2': 0}
n = X_train.shape[0]
r2_list = []
adjusted_r2_list = []

while True:
    changed = False
    print('')
    excluded = list(set(X.columns) - set(included))
    
    for col in excluded:
        pipe.fit(X_train[included + [col]], y_train)
        r2 = pipe.score(X_train[included + [col]], y_train)        
        k = len(included + [col])
        ar2 = 1 - (((1 - r2)*(n -1)) / (n - k - 1))
        
        if ar2 > best['ar2']:
            best = {'feature': col, 'r2': r2, 'ar2': ar2}            
            print(f'(B) Best! Feature: {best["feature"]}, R2: {best["r2"]:.3f}, Adjusted R2: {best["ar2"]:.3f}')
            changed=True
    if changed:
        included.append(best['feature'])
        r2_list.append(best['r2'])
        adjusted_r2_list.append(best['ar2'])
        print(f'Added {best["feature"]}')
    else:
        break

print(f'Features: {included}')
```

```python
plt.figure(figsize=(10, 6))
ranger = range(1, len(r2_list)+1)

plt.plot(ranger, r2_list, label = '$R^2$')
plt.plot(ranger, adjusted_r2_list, label = '$Adjusted \: R^2$')
plt.xlabel('Number of Features')
plt.legend()
plt.show()
```

#### Both Ridge & Lasso Regression resulted 60% accuracy. If simplicity is the question, we can use 'engine_hp', 'city_mpg', 'year', 'number_of_doors', 'engine_cylinders' and 'highway_mpg' as our imputs for our model with reduce accuracy of 56%

# Apply QLattice Regressor

```python
carz = car.copy()
```

```python
carz.info()
```

```python
carz.isnull().sum()
```

```python
carz = carz.drop(['make', 'model'], axis=1)
```

```python
# Drop values that are further than 3 STD from the mean
z = np.abs(stats.zscore(carz.price))
print(carz.shape[0])
```

```python
carz = carz[z < 3]
```

```python
carz.shape[0]
```

```python
from scipy import stats
import feyn
```

### QLattice Modelling

```python
train, test = train_test_split(carz, random_state=42)
target = 'price'
```

```python
# API is required for instantiate QLattice
qlattice = feyn.QLattice(url='your_qlattice_url', api_token='your_api_token')
```

```python
semantic_types = {}

for col in carz.columns:
    if carz[col].dtypes == 'object':
        semantic_types[col] = 'c'
```

```python
semantic_types
```

```python
qgraph = qlattice.get_regressor(train.columns, output = target, stypes = semantic_types,)
```

```python
qgraph.head(3)
```

```python
qgraph.best()
```

```python
no_loops = 10
for _ in range(no_loops):
    qgraph.fit(train, loss_function=feyn.losses.squared_error, threads = 6)
    best_graphs = qgraph.best()
    qlattice.update(best_graphs)
```

```python
best_graph = best_graphs[0]
best_graph
```

```python
prediction = best_graph.predict(test)
```

```python
predictions = prediction.round()
predictions
```

```python
sns.set_style('darkgrid')
feyn.plots.plot_regression_metrics(y_true = test[target], y_pred = predictions, 
                                   title = 'Evaluation Results')
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, sharex=True, figsize=(15,10))

ax1.scatter(test["engine_hp"],test["city_mpg"], c=test[target], s=5)
ax1.set_title('Actual price')
ax2.scatter(test["engine_hp"], test["city_mpg"],c=predictions, s=5)
ax2.set_title('Predicted price')
fig.suptitle('Engine HP vs City MPG')
```

```python
print(f'Base Model RMSE: {np.sqrt(mean_squared_error(y_test, y_pred))}')
print(f'QLattice RMSE: {np.sqrt(mean_squared_error(test[target], predictions))}')
```

```python
print(f'Base Model Score: {100*r2_score(y_test, y_pred):.2f}%')
print(f'Base Model Score: {100*r2_score(test[target], predictions):.2f}%')
```
